````python
def find_teacher(sub):
	Subjects = {'홍길동':['수학','과학'],'최영희':['영어', '수학'],
		'강동원':'영어', '정필수':['사회', ' 역사'],
		'박희수':'국어', '이승철':['수학', '과학']}
	teacher = []
	for key in Subjects.keys() :
		if sub in Subjects[key] :
			teacher.append(key)

	return teacher

sub = input('과목을 입력하시오: ')
print(find_teacher(sub))
````

### 변수

- 변수란?
    - 무엇을 담는 공간 혹은 영역입니다. 
    - 여기서 **`무엇`**은 **`데이터`**이고, **`공간 혹은 영역`**은 **`메모리`**를 말합니다. 
    - 즉, 프로그램에서 어떠한 데이터를 다루고 싶다면 꼭 변수를 사용해야 합니다. 
    

- 자료형이란?
    * 사람이 다루는 데이터는 다양한 형태이지만 파이썬의 변수는 몇 개의 유형으로 형태가 미리 정해져있습니다. 
    * 이러한 변수의 유형을 **자료형**이라고 합니다. 
    * 숫자 자료형은 모든 숫자를 표현할 수 있고,
    * 문자 자료형은 모든 문자열을 표현할 수 있고,
    * 리스트와 딕셔너리 자료형은 여러 문자열과 숫자를 혼합하여 표현할 수 있다.
    
    
- 변수 사용법

    - 변수는 '변할 수 있는 자료'로 프로그램에서 필요할 때 쓰기위해서 저장해 놓는 식별자이다.
    - 변수이름은 문자열로 된 것. 숫자로 시작x, 키워드X (예. name, str_data) (안되는 예. 1abc, print)
    - 모든 자료형 값을 다 저장할 수 있다.
    - 사용방법: 변수명 = 값

### Shallow copy와 Deep copy
* Shallow copy : 새로운 복합 객체를 만들고,(가능한 범위까지) 원본 객체를 가리키는 참조를 새로운 복합 객체에 삽입합니다.
* Deep copy : 새로운 복합 객체를 만들고,재귀적으로 원본 객체의 사본을 새로 만든 복합 객체에 삽입합니다.
* 참고 할 점 : 얕은 복사와 깊은 복사의 차이점은 복합 객체(리스트 또는 클래스 인스턴스들과 같은 다른 객체를 포함한 객체)에만 유효합니다.

````python
import copy

ori = [1,2]
a = [ori, 3]
b = copy.copy(a)
c = copy.deepcopy(a)
````

````python
b
````

````python
c
````

Shallow copy, Shallow copy 복사대상은 새로운 공간에 할당되지만 대상안에 있는 요소는 원본과 동일 대상을 참조합니다

````python
import copy

class MyClass:
    def __init__(self):
        self.listValue = [1,2]
        self.tupleValue = (1,2)

a = MyClass()
b = copy.copy(a)
c = copy.deepcopy(a)
````

Shallow copy
class안에 있는 mutable객체인 listValue는 원본과 동일 대상을 참조합니다.

Deep copy
class안에 있는 mutable객체인 listValue의 대한 새로운 사본을 만듭니다.
---------------------------------------------------------------------------

**요약**

1 deep copy target이 mutable 일 경우
- target : 원본과 같은 참조
- target의 포함된 객체 중 mutable 객체는 새로운 사본을 생성
- target의 포함된 객체 중 immutable 객체는 원본과 같은 참조

2 deep copy target이 immutable 일 경우
- target의 포함된 객체 중 mutable 객체가 있으면 target은 새로운 사본을 생성
- target의 포함된 객체 중 mutable객체가 없다면 target은 원본과 같은 참조

3 shallow copy target이 mutable 일 경우
- target : 새로운 사본을 생성
- target의 포함된 객체 : 원본과 같은 참조

4 shallow copy target이 immutable 일 경우
- target : 원본과 같은 참조
- target의 포함된 객체 : 원본과 같은 참조

5 immutable로만 구성된 객체일 경우
- 수정 불가하므로 새로운 공간을 만들지 않는 것으로 예상하며, =(Assignment)와 같게 동작합니다.
- deep, shallow copy 동일 함.

6 사용자 class 같은 개념으로 적용됨.

````python
def factorial(n) :
	print(f"함수 호출 n = {n}")
	if n == 1 :    # 종료 조건
		return 1
	else :
		res = n * factorial(n-1)
		print(f"중간 계산값 {n} * factorial({n-1}) = {res}")
	return res

i = 5 #@param{type:"integer"}
print(f"최종 결과값 factorial(i) = {factorial(i)}")
````

````python
def fibo(n) :
	if n == 1 or n == 2 :
		return 1
	else :
		return fibo(n-1) + fibo(n-2)

for i in range(1,15) :
	print(fibo(i))
````

````python
input_arr = [ 1,2,3,4,5 ]

def f(x):
    print(f"multiply 1000 {x}")
    return x * 1000

for i in map(f, input_arr):
    print(i)
````

결과값이 의도한 [ 1000, 2000, 3000, 4000, 5000 ]이 나오지 않았습니다.  
이는 아직 map이 평가(evaluate)되지 않아서 인데요,

아래와 같이 map을 실제 연산된 리스트로 변환합니다.

````python
input_arr = [ 1,2,3,4,5 ]
f = lambda x : x * 1000

[ *map(f, input_arr) ]
````

````python
for e in map(f, input_arr): 
    print(e)
````

## filter (데이터만 추려낼 때 사용)

---
파이썬의 내장 함수인 filter()는 여러 개의 데이터로 부터 일부의 데이터만 추려낼 때 사용합니다.  
따라서, 여러 개의 데이터를 담고 있는 list나 tuple을 대상으로 주로 사용하는 함수입니다.  

사용방법은 아래와 같습니다.  

filter(조건 함수, 순회 가능한 데이터)  
filter() 함수는 두번째 인자로 넘어온 데이터 중에서 첫번째 인자로 넘어온 조건 함수를 만족하는 데이터만을 반환합니다.  

map이 모든 원소에 대해 변환 연산을 수행한것과는 차이가 있습니다.

````python
input_arr = [ 1,2,3,4,5,6 ]
f = lambda x : x % 2 == 0

for x in filter(f, input_arr):
    print(x)
````

위와 같이 x % 2가 0인, 즉 짝수인 원소들만 추리고 싶을때에 위와 같이 filter함수를 사용합니다.  
조건식 또한 다양하게 적용할 수 있어 매우 많이 사용되는 함수입니다.

--- 
input_arr이 [ 1, -4, 9, 12, -15 ] 라고 할 때 원소의 제곱이 81이상인 원소만 추리고자 합니다.  
filter 함수를 활용하여 해결해 봅시다.  

````python
input_arr = [ 1, -4, 9, 12, -15 ]
f = lambda x: (x ** 2) >= 81
[ *filter(f, input_arr) ] 
````

---
input_arr 의 모든 원소의 절대값에서 10을 빼주고, 이 결과가 0 이상인 원소들만 추리고자 합니다.  
map 함수와 filter 함수를 조합하여 해결해 봅시다.  

````python
input_arr = [ 1, -4, 9, 12, -15 ]

f0 = lambda x : abs(x) - 10
minus_10 = map(f0, input_arr)

f1 = lambda x : x > 0

[*filter(f1, minus_10) ]
````

## reduce (요소와 연산을 하나의 변수로)

배열의 원소를 하나씩 방문해 특정 연산 결과를 하나의 변수의 누적하고자 할 때 사용되는 함수

````python
from functools import reduce

arr = [1,2,3,4,5]
add = lambda acc, x: acc + x

reduce(add,arr)
````

위와 같이 arr을 하나씩 순회하며 add 함수를 수행한 연산 결과를 다음 인자값으로 전달합니다.  

---

````python
reduce(add,arr,100)
````

위와 같이 초기값을 정의할 수도 있습니다.  
위코드는 초기값 100이라는 값에 1을 더한 후 101을 다음 acc 인자값으로 전달하여 줍니다.  
그리고 101 + 2 의 결과인 103을 다음 acc 인자값으로 전달하는 연산을 배열 순회가 종료 될때까지 반복 합니다.  
---

--- 

어떤 기업은 아래와 같은 연봉 시스템이 있다고 합니다.

1. 초봉은 6000만원 이며
2. 1년 마다 5%의 연봉 인상이 있으며
3. 5년 주기로 4%의 추가 인상이 있습니다.
4. 20년 이상의 연차에서는 연봉 인상이 없어지고 매년 2% 씩 삭감된다고 합니다.
5. 20년 이상의 경우 5년주기의 추가 인상분은 없습니다.

만약 이 기업에 입사하여 25년을 다닌다면 25년 째되는 마지막 급여는 얼마일까요?

````python
def cal(before_salary, year):
    now_salary = before_salary
    if year < 20 :
        now_salary *= 1.05 
    
        if year % 5 == 0 :
            now_salary *= 1.04
    
    else:
        now_salary *= 0.98
    print(year, now_salary)    
    return now_salary

reduce(cal, range(1, 26), 6000)
    
````

````python
profits = [ 1.04, 1.02, 1.001, 0.9, 0.94, 1.01, 0.9 ]
seed = 10000

def f(money, profit):
    print(money, profit, money * profit)
    return money * profit

reduce(f, profits, seed)
````

````python
# 여러가지 색 정사각형 그리기, 함수

def square(t, size, color):
	t.color(color)
	for i in range(4):
		t.forward(size)
		t.right(90)

t.clearscreen()
t.speed(13)
t.pensize(3)
colors = ['red', 'orange', 'yellow', 'green', 'blue', 'violet']

i = 30
for color in colors:
	square(t, i, color)
	i = i + 30
````

````python
# 다각형 그리기, 함수

def drawPolygon(t, sideLength, numSides, color):
	t.color(color)
	turnAngle= 360 / numSides
	for i in range(numSides):
		t.pendown()
		t.forward(sideLength)
		t.right(turnAngle)

t.clearscreen()
t.speed(13)
for i in range(3):
	t.penup()
	t.setposition(40*i, 0)
	drawPolygon(t, 20, 6, "blue")

for i in range(5):
	t.penup()
	t.setposition(40*(i-2), -100)
	drawPolygon(t, 40, 4, "violet")
````

````python
# 꽃 그리기, 함수

def flower(t, n, r):
    angle = 360/n
    for i in range(n):
        for i in range(2):
            t.circle(r,angle)
            t.left(180-angle)
        t.left(angle)

def move(t, length):
	t.penup()
	t.forward(length)
	t.pendown()

t.clearscreen()
t.speed(13)
t.color("violet")
move(t, -100)

for i in range(3):
    flower(t, 6, 30+(50*i))
````

````python
# 복잡한 예제1 (문양 그리기)

def design(homePos, scale):
    t.left(18)
    t.up()
    for i in range(5):
        t.forward(64.65 * scale)
        t.down()
        wheel(t.position(), scale)
        t.up()
        t.backward(64.65 * scale)
        t.right(72)
    t.up()
    t.goto(homePos)
    t.right(36)
    t.forward(24.5 * scale)
    t.right(198)
    t.down()
    centerpiece(46 * scale, 143.4, scale)
    t.done()

def wheel(initpos, scale):
    t.right(54)
    for i in range(4):
        pentpiece(initpos, scale)
    t.down()
    t.left(36)
    for i in range(5):
        tripiece(initpos, scale)
    t.left(36)
    for i in range(5):
        t.down()
        t.right(72)
        t.forward(28 * scale)
        t.up()
        t.backward(28 * scale)
    t.left(54)
    t.done()

def tripiece(initpos, scale):
    t.pencolor("black")
    oldh = t.heading()
    t.down()
    t.backward(2.5 * scale)
    tripolyr(31.5 * scale, scale)
    t.up()
    t.goto(initpos)
    t.setheading(oldh)
    t.down()
    t.backward(2.5 * scale)
    tripolyl(31.5 * scale, scale)
    t.up()
    t.goto(initpos)
    t.setheading(oldh)
    t.left(72)
    t.done()

def pentpiece(initpos, scale):
    t.pencolor("black")
    oldh = t.heading()
    t.up()
    t.forward(29 * scale)
    t.down()
    for i in range(5):
        t.forward(18 * scale)
        t.right(72)
    pentr(18 * scale, 75, scale)
    t.up()
    t.goto(initpos)
    t.setheading(oldh)
    t.forward(29 * scale)
    t.down()
    for i in range(5):
        t.forward(18 * scale)
        t.right(72)
    pentl(18 * scale, 75, scale)
    t.up()
    t.goto(initpos)
    t.setheading(oldh)
    t.left(72)
    t.done()

def pentl(side, ang, scale):
    t.pencolor("green")
    if side < (2 * scale): return
    t.forward(side)
    t.left(ang)
    pentl(side - (.38 * scale), ang, scale)

def pentr(side, ang, scale):
    t.pencolor("green")
    if side < (2 * scale): return
    t.forward(side)
    t.right(ang)
    pentr(side - (.38 * scale), ang, scale)

def tripolyr(side, scale):
    t.pencolor("red")
    if side < (4 * scale): return
    t.forward(side)
    t.right(111)
    t.forward(side / 1.78)
    t.right(111)
    t.forward(side / 1.3)
    t.right(146)
    tripolyr(side * .75, scale)

def tripolyl(side, scale):
    t.pencolor("red")
    if side < (4 * scale): return
    t.forward(side)
    t.left(111)
    t.forward(side / 1.78)
    t.left(111)
    t.forward(side / 1.3)
    t.left(146)
    tripolyl(side * .75, scale)

def centerpiece(s, a, scale):
    t.pencolor("red")
    t.forward(s); t.left(a)
    if s < (7.5 * scale):
        return
    centerpiece(s - (1.2 * scale), a, scale)

import math
#import ColabTurtlePlus.Turtle as t 

t.clearscreen()
t.initializeTurtle(window=(550,550))
t.pencolor("black")
t.pensize(1)
t.animationOff()
t.speed(0)
t.hideturtle()
design(t.position(), 2)
t.done()
t.saveSVG("bytedesign-colored",turtle=False)
````

````python
# 복잡한 예제2 (거북이 경주)

#import ColabTurtlePlus.Turtle as t 
import random
from time import sleep

# Screen setup
t.clearscreen()

t.setup(800,500)
t.bgcolor("forestgreen")
t.speed(0)  # Turn off drawings until done()

# Heading
t.penup()
t.goto(0,215)
t.color("white")
t.write("TURTLE RACE", align="center", font=("arial", 20, "bold"))

# Dirt
t.goto(-350,200)
t.pendown()
t.color("saddlebrown")
t.begin_fill()
for i in range(2):
  t.forward(700)
  t.right(90)
  t.forward(400)
  t.right(90)
t.end_fill()

# Finish Line
gap_size = 20
t.shape("square")
t.penup()
t.color("white")
for i in range(10):
  t.goto(250,(170-(i*gap_size*2)))
  t.stamp(layer=1)
for i in range(10):
  t.goto(250+gap_size,(210-gap_size-(i*gap_size*2)))
  t.stamp(layer=1)
t.color("black")
for i in range(10):
  t.goto(250,(190-(i*gap_size*2)))
  t.stamp(layer=1)
for i in range(10):
  t.goto(251+gap_size,(190-gap_size-(i*gap_size*2)))
  t.stamp(layer=1)
t.done()

# Create the racing turtles
turtles = [t.Turtle(), t.Turtle(), t.Turtle(), t.Turtle()]
for i in range(4): 
    turtles[i].hideturtle()
colors = ["cyan", "magenta", "yellow", "lime"]
for i in range(4):
  turtles[i].shape("turtle2")  
  turtles[i].color(colors[i])
  turtles[i].shapesize(1.5)
  turtles[i].speed(10)
  turtles[i].showturtle()
  turtles[i].penup()
  turtles[i].goto(-300,150-100*i)
  turtles[i].pendown()

# Pause for 1 second before starting race
sleep(1)

# The race
random.seed()
positions = [turtles[i].xcor() for i in range(4)]
while max(positions) <= 230:
  turtles[0].forward(random.randint(1,10))
  turtles[1].forward(random.randint(1,10))
  turtles[2].forward(random.randint(1,10))
  turtles[3].forward(random.randint(1,10))
  positions = [turtles[i].xcor() for i in range(4)]

# Determine the winner with possibility of a tie
win_pos = max(positions)
winners = [i for i, j in enumerate(positions) if j == win_pos]
for k in winners:
  turtles[k].speed(5)
  turtles[k].left(360)

# Print message about who won
t.hideturtle()
t.color("white")
t.goto(0, -235)
if len(winners)==1:
  t.write("Turtle "+colors[winners[0]]+" won!", align="center", font=("Arial", 20, "bold"))
else:
  s = colors[winners[0]]
  for k in range(1,len(winners)):
    s += " and " + colors[winners[k]]
  t.write("Turtles "+s+ " tied!", align="center", font=("Arial", 20, "bold"))
t.done()

# Print the final positions for each turtle
pad = max([len(s) for s in colors])+10
for k in range(4):
  print(("turtle "+colors[k]).ljust(pad), int(positions[k]))
````

````python
# 복잡한 예제3 (그래프 그리기)
from math import *

def draw_function(t, formula, x_width):
    t.pencolor('red')
    t.pensize(3)

    # 그래프 그리는 초기 위치로 이동
    t.penup()
    magnify = axis_size / x_width  # 정의역에 따른 그래프 확대 배율
    initial_x = -x_width    # x 좌표 초기값
    x = initial_x   # eval 함수를 쓰기 위해 x 에 x 좌표 초기값을 입력
    y = eval(formula)
    t.goto(magnify*x, magnify*y)
    t.pendown()

    resolution = 100    # 해상도, 그래프 위의 점의 개수
    for n in range(resolution+1):
        x = initial_x * (1 - 2*n/resolution)    # 증가하는 x좌표값
        try: y = eval(formula)
        except: continue
        t.goto(magnify*x, magnify*y)

    # 좌표값 입력
    t.pencolor('black')
    t.pensize(1)
    t.penup()
    t.goto(-axis_size, -20)
    t.write(initial_x)
    t.goto(-20, -20)
    t.write(0)
    t.goto(axis_size-10, -20)
    t.write(-initial_x)
    t.goto(-20, axis_size-20)
    t.write(-initial_x)
    t.goto(-20, -axis_size+20)
    t.write(initial_x)
    
# 좌표축 그리기
def draw_axis(t, axis_size):
    for i in range(4):
        t.setheading(90 * i)
        t.forward(axis_size)
        t.home()

# 화면 지우기
def clear(t, speed, axis_size):
    t.reset()
    t.speed(speed)
    t.hideturtle()
    draw_axis(t, axis_size)

# Screen setup
t.clearscreen()

# 기본 설정
axis_size = 300
speed = 0
clear(t, speed, axis_size)

#@title 함수의 정의 
#@markdown 예시) sin(x), 곱하기는 생략 불가
함수 = "sin(x)" #@param{type:"string"}

# 정의역의 범위
배율 =  4#@param{type:"integer"}

# 그래프 그리기
draw_function(t, 함수, 배율)
t.done()
````
